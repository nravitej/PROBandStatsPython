```python
import numpy as np
import math as mt
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import collections 
import sys

sample_size=input("Enter sample size \n")
samples=input("Enter samples")
TossCount=input("Enter Toss Count=>N \n")
Coinsinhand=input("Coins in hand for single Toss \n")


try:
    sample_size=int(sample_size)
    samples=int(samples)
    TossCount=int(TossCount)
    Coinsinhand=int(Coinsinhand)
except ValueError:
    print("Kindly input the proper values")
    sys.exit(1)

l1=np.random.randint(low=0,high=Coinsinhand+1,size=TossCount) #CoinToss Population
coins=pd.DataFrame(l1)
    
l2=[np.sum((coins.sample(n=sample_size)).values) for i in range(samples)] #List Comprehension #List of Means of samples      
    
l2=np.array(l2)/sample_size

l2.sort()
p=collections.Counter(l2)
x=[key for key in p]
y=[p[key] for key in p]
pmax=np.max(np.array(y))
y=[i*(2) for i in y]
kde = stats.gaussian_kde(l2)
pos = np.linspace(np.min(l2), np.max(l2), 10000)
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.xlim(0,Coinsinhand)
plt.xlabel('Mean distribution of Toss Samples')
plt.ylabel('Occurunce of Heads')
plt.title('Illustrating Central Limit Theorem for Coin Toss')
t=plt.hist(l2,bins=30)
plt.plot(pos, kde(pos)*pmax*(Coinsinhand/2), color='red')

plt.subplot(1,2,2)
plt.xlabel('Mean distribution of Toss Samples')
plt.ylabel('Occurunce of Heads')
plt.title('The hist curve in corelation with Normal distribution Curve')

plt.plot(x,y)
plt.plot(pos, kde(pos)*pmax*(Coinsinhand/2), color='red')

```
